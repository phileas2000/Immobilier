```python
#Importation data(test)
data.to_sql('Data',conn, if_exists='replace')
print(pd.read_sql_query("SELECT * FROM Data",conn))
```

```python
print(pd.read_sql_query("SELECT \"Date mutation\" FROM Data WHERE strftime('%m', \"Date mutation\") > 06 ORDER BY \"Date mutation\"",conn))
```

```python
print(pd.read_sql_query("SELECT \"Date mutation\" FROM Data ORDER BY \"Date mutation\"",conn))
cursor.execute("DELETE FROM Data WHERE strftime('%m', \"Date mutation\") > 06 ")
conn.commit()
print(pd.read_sql_query("SELECT \"Date mutation\"  FROM Data ORDER BY \"Date mutation\" ",conn))
```

```python
data = pd.read_sql_query("SELECT * FROM Data",conn)
print(data)
```

```python
data['Id'] = range(len(data))
print(data['Id'] )
```

```python
#Importation type local
typeLocal = pd.DataFrame()
typeLocal[['Type local','Code type local']] = data[['Type local','Code type local']].drop_duplicates(subset=['Type local'])
typeLocal.to_sql('Type local',conn, if_exists='replace')
print(pd.read_sql_query("SELECT * FROM \"Type local\"",conn))
```

```python
#Importation mutation
mutation = pd.DataFrame()
mutation[['Date mutation','Nature mutation','Valeur fonciere','Id']] = data[['Date mutation','Nature mutation','Valeur fonciere','Id']]
mutation.to_sql('Mutation',conn, if_exists='replace',index = False)
print(pd.read_sql_query("SELECT * FROM Mutation",conn))
```

```python
#Importation batiment
batiment = data.drop(columns = ['Date mutation','Nature mutation','Valeur fonciere','Type local','Code type local','Date mutation','Nature mutation','Valeur fonciere'])
batiment = batiment.rename(columns={'Id': 'IdMutation'})
batiment.to_sql('Batiment',conn, if_exists='replace')
print(pd.read_sql_query("SELECT * FROM Batiment",conn))
```
